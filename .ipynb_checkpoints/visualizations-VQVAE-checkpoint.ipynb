```python
# Modified from code by (c) 2021 Rui Shu
import argparse
import numpy as np
import torch
import tqdm
from codebase import utils as ut
from codebase.models.vae import VAE
from codebase.models.gmvae import GMVAE
from pprint import pprint
import torchvision
from torchvision import datasets, transforms
import matplotlib.pyplot as plt
```

```python
#args and setup
iter_max = 20000
z_dim = 10
k = 500 #for GMVAE
run = 0

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
_, labeled_subset, _ = ut.get_mnist_data(device, use_test_subset=True)
x,y = labeled_subset
```

```python
def plot_grid_images(xhat, model_name):
    xhat = xhat.view(-1,1,28,28) # N x C x H x W
    # make grid to display images
    grid_img = torchvision.utils.make_grid(xhat, nrow=20)
#     print(grid_img.shape)
    # reshape and plot (because MPL needs channel as the last dimension)
    grid_img = grid_img.permute(1, 2, 0).detach().numpy()
    plt.figure(figsize = (20,10))
    plt.imshow(grid_img)
    plt.title(f'Model: {model_name}, Num Samples: {xhat.shape[0]}')
    plt.show()
```

## VAE

```python
layout = [
    ('model={:s}',  'vae'),
    ('z={:02d}',  z_dim),
    ('run={:04d}', run)
]
model_name = '_'.join([t.format(v) for (t, v) in layout])

print('Model name:', model_name)
vae = VAE(z_dim=z_dim, name=model_name).to(device)
ut.load_model_by_name(vae, global_step=iter_max, device=device)
```

```python
# generate images sampled from z-space
num_images = 200
xhat = vae.sample_x(num_images)

# plot sample images
plot_grid_images(xhat, model_name)
```

## GMVAE

```python
run = 1
layout = [
    ('model={:s}',  'gmvae'),
    ('z={:02d}',  z_dim),
    ('k={:03d}',  k),
    ('run={:04d}', run)
]
model_name = '_'.join([t.format(v) for (t, v) in layout])

print('Model name:', model_name)
gmvae = GMVAE(z_dim=z_dim, k=k, name=model_name).to(device)
ut.load_model_by_name(gmvae, global_step=iter_max, device=device)
```

```python
# generate images sampled from z-space
num_images = 200
xhat = gmvae.sample_x(num_images)

# plot sample images
plot_grid_images(xhat, model_name)
```
